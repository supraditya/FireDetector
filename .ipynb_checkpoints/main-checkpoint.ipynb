```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import keras
from keras.models import Sequential,Model
from keras.optimizers import Adam
from keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, GlobalMaxPooling2D
from keras.utils.np_utils import to_categorical
```

```python
x=list()
y=list()
for i in range(1,111):
    img=cv2.imread("New Fire Images/%d.jpg"%i,1)
    img=cv2.resize(img,(256,256))
    #img=np.float32(img)
    #img/=255.0
    img = img.reshape(256, 256, 3)
    img=img.tolist()
    x.append(img)
    y.append(1)
for i in range(1,111):
    img=cv2.imread("new normal/%d.jpg"%i,1)
    img=cv2.resize(img,(256,256))
    #img=np.float32(img)
    #img/=255.0
    img = img.reshape(256, 256, 3)
    img=img.tolist()
    x.append(img)
    y.append(0)
x=np.asarray(x)
y=np.asarray(y)
y=to_categorical(y)
```

```python
def NN():
    model = Sequential()
    model.add(Conv2D(16, (5,5),input_shape=(256,256,3),padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Conv2D(8, (3, 3), activation = 'relu',padding = 'same'))
    model.add(MaxPooling2D(pool_size = (2,2)))
    model.add(Flatten())
    model.add(Dense(1024, activation = 'relu'))
    model.add(Dense(512, activation = 'relu'))
    model.add(Dense(2, activation = 'softmax'))
    model.compile(Adam(lr = 0.01), loss = 'categorical_crossentropy', metrics = ['accuracy'])
    return model
```

```python
model=NN()
model.summary()
```

```python
model.fit(x,y,epochs=10,batch_size=5,verbose=1,shuffle=True,validation_split=0.1)
```
